```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/2.csv')
df.head()
```

```python
df.describe()
```

```python
df.isna().sum()
```

```python
data_1 = df['math score']
data_2 = df['reading score']
data_3 = df['writing score']
data = [data_1, data_2, data_3]
 
fig = plt.figure(figsize =(10, 7))
 
# Creating axes instance
ax = fig.add_axes([0, 0, 1, 1])

x = [1, 2, 3]
# Creating plot
bp = ax.boxplot(data)

labels = ['math score', 'reading score', 'writing score']
plt.xticks(x, labels, rotation ='vertical')
 
# show plot
plt.show()
```

```python
df['math score'] = df['math score'] / 10
df['reading score'] = df['reading score'] / 10
df['writing score'] = df['writing score'] / 10
```
